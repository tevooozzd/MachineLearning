```python
import numpy as np
vetorA = np.array([1,2,3]) # 1x3
vetorB = np.array([4,5,6]) # 1x3 
vetorC = np.array([7,8,9]) # 1x3
print(f"Vetores A, B e C criados, e possuem respectivamente os valores:{vetorA}, {vetorB} e {vetorC}")

# Calculando o vetor médio
vetor_medio = (vetorA + vetorB + vetorC) / 3
print(f"O vetor médio é: {vetor_medio}")
```
