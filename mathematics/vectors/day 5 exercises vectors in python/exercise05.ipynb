```python
import numpy as np
vetorA = np.array([2,3,1]) # 1x3
vetorB = np.array([-3,2,1]) # 1x3 
print(f"Vetores A e B criados, e possuem respectivamente os valores:{vetorA} e {vetorB}")

# Calculando o produto escalar (A * B)
produto_escalar = np.dot(vetorA, vetorB)
print(f"O produto escalar dos vetores A e B é: {produto_escalar}")

# Vendo se são perpendiculares
magnitudeA = np.linalg.norm(vetorA)
magnitudeB = np.linalg.norm(vetorB)
print(f"A magnitude dos vetores A e B é respectivamente: {magnitudeA:.2f}, {magnitudeB:.2f}")

# Achando o cosseno
cosseno0 = produto_escalar / (magnitudeA * magnitudeB)
print(f"O cosseno é: {cosseno0:.2f}")

# Transformando para radiando e depois para graus
radiando = np.arccos(cosseno0)
graus = np.degrees(radiando)
print(f"O ângulo dos vetores A e B é: {graus:.2f}")

if np.isclose(produto_escalar, 0, atol=1e-5) and np.isclose(graus, 90, atol=1e-2):
    print(f"Os vetores A e B SÃO perpendiculares, pois têm produto escalar ≈ {produto_escalar} e ângulo ≈ {graus:.2f}°")
else:
    print(f"Os vetores A e B NÃO são perpendiculares, pois têm produto escalar = {produto_escalar} e ângulo = {graus:.2f}°")
```
