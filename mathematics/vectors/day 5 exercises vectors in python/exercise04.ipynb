```python
import numpy as np
vetorA = np.array([4,1,-2]) # 1x3
vetorB = np.array([1,2,3]) # 1x3 
print(f"Vetores A e B criados, e possuem respectivamente os valores:{vetorA} e {vetorB}")

# Calculando o produto escalar (A * B)
produto_escalar = np.dot(vetorA, vetorB)
print(f"O produto escalar dos vetores A e B é: {produto_escalar}")

# Calculando a magnitude/normalizando
magnitudeA = np.linalg.norm(vetorA)
magnitudeB = np.linalg.norm(vetorB)
print(f"A magnitude dos vetores A e B é respectivamente: {magnitudeA:.3f}, {magnitudeB:.3f}")

# Achando o cosseno
cosseno0 = produto_escalar / (magnitudeA * magnitudeB)
print(f"O cosseno é: {cosseno0}")

# Transformando para radiando e depois para graus
radiando = np.arccos(cosseno0)
graus = np.degrees(radiando)
print(f"O ângulo dos vetores A e B é: {graus}")
```
