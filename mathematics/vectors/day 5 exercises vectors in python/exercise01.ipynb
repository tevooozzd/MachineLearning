# Manipulando vetores

```python
# Criando os vetores
import numpy as np
vetorA = np.array([3,-2,1]) # 1x3
vetorB = np.array([4,0]) # 1x2 
print(f"Vetores A e B criados, e possuem respectivamente os valores:{vetorA} e {vetorB}")

# Verificando se podem ser manipulados

if np.shape(vetorA) == np.shape(vetorB):
    print("Podemos continuar!")
    vetorC = vetorA + vetorB
    print(f"O novo Vetor C criado pela soma de A + B é:\n{vetorC}")
else:
    print(f"!ALERTA! Algo está errado e devemos arrumar as dimensões")
    vetorB = np.append(vetorB, 3) # Adicinado mais um elemento (index 2) ao vetor B // AGora 1x3
    print(f"!ALERTA! Agora sim, tudo certo, novo número adicionado e ele foi o 10!")
    vetorC = vetorA + vetorB
    print(f"O novo Vetor C criado pela soma de A + B é:{vetorC}")
```
