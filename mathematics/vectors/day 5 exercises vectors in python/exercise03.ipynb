```python
import numpy as np
vetorA = np.array([6,8]) # 1x2
print(f"Vetor A criado, e possue o valor de:{vetorA}")

# Calculando a normalização/magnitude 
normalizacao_magnitude = np.linalg.norm(vetorA)

# Verificando se é unitario, ou seja magnitude == 1
if normalizacao_magnitude != 1:
    normalizado = vetorA / normalizacao_magnitude
    print(f"O novo vetor normalizado (magnitude 1) é: {normalizado}")
else:
    print(f"O vetor é vetor unitário, ou seja magnitude igual a: {normalizacao_magnitude}")
```
