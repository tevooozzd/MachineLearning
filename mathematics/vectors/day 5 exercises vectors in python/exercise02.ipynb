```python
import numpy as np
vetorA = np.array([3,1]) # 1x2
vetorB = np.array([-1,3]) # 1x2 
print(f"Vetores A e B criados, e possuem respectivamente os valores:{vetorA} e {vetorB}")

# Calculando o produto escalar (A * B)
produto_escalar = np.dot(vetorA, vetorB)

# Verificando se são ortogonais, ou seja produto escalar == 0
if produto_escalar:
    print(f"O produto escalar dos vetores A e B é: {produto_escalar}, portanto não é escalar")
else:
    print(f"O produto escalar dos vetores A e B é: {produto_escalar}, portanto é escalar")
```
